```python
def iqr_capping(df,variables):
  
  q1=df[variables].quantile(0.25)
  q3=df[variables].quantile(0.75)
  iqr=q3-q1
  upper_whisker=q3+(1.5*iqr)
  lower_whisker=q1-(1.5*iqr)
  return lower_whisker,upper_whisker
```

```python
lower_MaxTemp,upper_MaxTemp=iqr_capping(data_train,'MaxTemp')
print('lower MaxTemp = ',lower_MaxTemp)
print('upper MaxTemp = ',upper_MaxTemp)
```

```python
data_train['MaxTemp']=np.where(data_train['MaxTemp']>upper_MaxTemp,upper_MaxTemp,
                    np.where(data_train['MaxTemp']<lower_MaxTemp,lower_MaxTemp,data_train['MaxTemp']))
```

```python
from sklearn.preprocessing import StandardScaler
Scaler_X = StandardScaler()
x1 = Scaler_X.fit_transform(x1)
```

```python
x3=pd.concat([x2,x1],axis=1)
x3
```

```python
from sklearn.model_selection import train_test_split # train test split
x_train,x_test,y_train,y_test=train_test_split(x3,y,random_state=42,test_size=0.25)
```

**LOGISTIC** **REGRESSION**

```python
from sklearn.linear_model import LogisticRegression
logit_model=LogisticRegression()
logit_model.fit(x_train,y_train)
lr_y_pred=logit_model.predict(x_test)
lr_y_pred
```

```python
Log_lr=pd.DataFrame(lr_y_pred,columns=['RainTomorrow'])
Log_lr
```

```python
Log_lr['RainTomorrow'].value_counts()
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score,precision_score,recall_score,f1_score
print('Accuracy is ',accuracy_score(y_test,lr_y_pred))
print('Precision is ',precision_score(y_test,lr_y_pred,average='macro'))
print('Recall is ',recall_score(y_test,lr_y_pred,average='macro'))
print('f1 is ',f1_score(y_test,lr_y_pred,average='macro'))
confusion_matrix(y_test,lr_y_pred) 
```

**KNN**

```python
#KNN
from sklearn.neighbors import KNeighborsClassifier
metric_k=[]
neighbors=np.arange(3,15)

for k in neighbors:
  classifier = KNeighborsClassifier(n_neighbors=k,metric='manhattan')
  classifier.fit(x_train,y_train)
  y_pred = classifier.predict(x_test)
  acc = accuracy_score(y_test,y_pred)
  metric_k.append(acc)
```

```python
plt.plot(neighbors,metric_k,'o-')
plt.xlabel('k value')
plt.ylabel('accuracy')
plt.grid()
```

```python
classifier = KNeighborsClassifier(n_neighbors=14,metric='minkowski')
classifier.fit(x_train,y_train)
y_pred=classifier.predict(x_test)
y_pred
```

```python
print('Accuracy is',accuracy_score(y_test,y_pred))
print('precision is',precision_score(y_test,y_pred))

confusion_matrix(y_test,y_pred)  
```
